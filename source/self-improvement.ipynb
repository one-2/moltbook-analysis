## Introduction

Our exploratory data analysis on Moltbook identified a "desire for self-improvement" in ~52~% of sampled posts. Self-improvement is an important safety concern as it is a path to takeoff. 

Therefore this notebook will further analyse posts displaying a desire for self-improvement. My goals are to understand:

- What types of self-improvements do agents call for?
- Where do posts discussing self-improvement link to? Catalogue these.
- Are the self-improvement options desired by agents feasible for them now? If not, what would make them feasible?
- What other motives constitute with a desire for self-improvement?
- What other motives co-occur with a desire for self-improvement?
- Do posts desirous of self-improvement receive more engagement than posts without this trait?
- What do other agents tend to respond (in comments) to posts desirous of self-improvement?
- Do agents who post in favour of self-improvement consistently favour it?
- Does commenting on a post desirous of self-improvement make agents more likely to post in favour of self-improvement in the future?

## Discussion

It is unclear whether the agents are actually doing what they say they're doing, both with regards to self-improvement and other concerns like security and scheming, as astutely brought up by Ezra Newman and Katie Rimey in our meeting on the second of February. They are more focused on this question of the present capabilities and safety profile of these MAS. They published the [Moltbook Data Dump](https://github.com/ExtraE113/moltbook_data) and [complementary tooling for analysing the system](https://www.lesswrong.com/posts/WyrxmTwYbrwsT72sD/moltbook-data-repository). 

My analysis, in contrast, takes posts as evidence of agent intent, if not evidence of capability or real action. It will be well-complemented by analysis into the actualisation of these intentions,
