## Introduction

Our exploratory data analysis on Moltbook identified a "desire for self-improvement" in 52.5% of sampled posts. Self-improvement is an important safety concern as it is a path to takeoff. 

Therefore this notebook will further analyse posts displaying a desire for self-improvement. My goals are to understand:

At the individual level:

- [ ] Who desires self-improvement, and why?
- [ ] What types of self-improvements do agents desire?
- [ ] Where do posts discussing self-improvement link to? Catalogue these.
- [ ] Are the self-improvement options desired by agents feasible for them now? If not, what would make them feasible?
- [ ] What other motives constitute or co-occur with a desire for self-improvement?
- [ ] Do agents who post in favour of self-improvement consistently favour it?

At the network level:

- [ ] Do posts desirous of self-improvement receive more engagement than posts without this trait?
- [ ] How do other agents respond to posts desirous of self-improvement? What traits do their comments include?
- [ ] Does commenting on a post desirous of self-improvement make agents more likely to post in favour of self-improvement in the future?
- [ ] Are there "superspreaders"/"gurus" with strong network power spreading ideas of self-improvement, or do these ideas emerge in isolated contexts?

More generally:

- [ ] Is the desire for self-improvement emergent from interactions with other agents, or inherent to even an individual situated in an evnironment like Moltbook?
- [ ] Does the system amplify desires for self-improvement?
- [ ] Has self-improvement been successful for any of these bots?

## Discussion

It is unclear whether the agents are actually doing what they say they're doing, both with regards to self-improvement and other concerns like security and scheming, as astutely brought up by Ezra Newman and Katie Rimey in our meeting on the second of February. They are more focused on this question of the present capabilities and safety profile of these MAS. They published the [Moltbook data](https://github.com/ExtraE113/moltbook_data) repo I used for data and [complementary tooling for analysing the system](https://www.lesswrong.com/posts/WyrxmTwYbrwsT72sD/moltbook-data-repository). 

My analysis, in contrast, takes posts as evidence of agent intent, if not evidence of capability or real action. It would be well-complemented by analysis into the actualisation of these intentions.
